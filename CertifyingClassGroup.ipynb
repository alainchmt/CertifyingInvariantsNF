```python
# Given a ring O in a number field K, 
# a prime ideal I above p, and an element b in O, this function finds the minimal polynomial of (b mod I). 

def FindMinimalPolynomial (K, O, I, p, b):
    R = QuotientRing(O , I, 'y')
    P = K(b).minpoly()
    Rp.<x> = PolynomialRing(GF(p))
    F = factor(Rp(P))
    for t in F:
        if ZZ[x](t[0])(R(b)) == 0:
            return t[0]
   
```

```python
# Given an irreducible polynomial over F_p, this function finds a primitive root over the field extension F_p[X]/(f).
# It returns a polynomial over F_p such that its reduction is a primitive root. 

def FindPrimitiveRoot (f, p):
    n = f.degree()
    R.<x> = PolynomialRing(GF(p))
    I = R.ideal(f)
    Q = QuotientRing(R, I)
    F = factor(p^n - 1)
    for i in range(10):
        set_random_seed(i)
        r = R.random_element()
        if r != 0 :
            flag = 0
            for t in F:
                if Q(r ^ ((p ^ (n)-1)/t[0])) == 1:
                    flag = 1
            if flag == 0:
                return r 
    
```

```python
# Given a ring O in a number field K, 
# a prime ideal I above p, and an element b in O which is primitive in O/I, 
#this function finds a primitive root for (O/I)* in terms of b

def FindPrimitiveRootInQuotient (K, O, I, p, b):
    f = FindMinimalPolynomial(K, O, I, p, b)
    R = QuotientRing(O , I, 'y')
    if f.degree() == int(log(I.norm())/log(p)):
        g = FindPrimitiveRoot(f, p)
        return (ZZ[x](g))(R(b))
    else :
        print('element is not primitive')
```

```python
# Given an element s in O \ I, and using a primitive root for (O/I)* computed using a primitive element b in O, 
# this function finds (m mod q), where q is a prime dividing N(I)-1.  

def QDiscreteLogarithm (K, O, I, p, b, s, q):
    R = QuotientRing(O , I, 'y')
    g = FindPrimitiveRootInQuotient (K, O, I, p, b)
    if s in I:
        print ('The element is in I')
    if (I.norm()-1) % q != 0 :
        print (f'Warning: {q} does not divide N(I)-1') 
    for i in range(I.norm()-1):
        if g ^ i == R(s):
            return (i % q)
```

```python
# Given a prime number p and an element s, these generates n prime ideals q such that p divides N(q)-1 and s is not in q. 

def PrimeIdealsLog (K, O, s, p, n):
    L = []
    cont = 0
    while len(L) < n or cont > 100:
        I = O.ideal(Primes()[cont])
        F = factor(I)
        for t in F:
            if (t[0].norm()-1) % p == 0 and (not (s in t[0])):
                L = L + [t[0]]
        cont = cont + 1
    return L[0 : n]
    
```

```python
#Given a list of units L, an element s, and a prime p, this generates an n x (len(L) + 1) matrix, where each row is obtain by applying 
# a different q-logarithm. 

def MatrixQLogarithm (K, O, L, s, p, n):
    Q = PrimeIdealsLog (K, O, s, p, n)
    L = L + [s]
    A = [[0 for j in range(len(L))] for i in range(n)]
    for i in range(n) :
        for j in range(len(L)):
            #print(Q[i])
            A[i][j] = QDiscreteLogarithm(K, O, Q[i], int(Q[i].gens()[0]), K(a+1), L[j], p)
    return Matrix(A)
```
