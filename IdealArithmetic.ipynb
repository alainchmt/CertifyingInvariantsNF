```python
K = NumberField(x^5 - 5*x^3 + 10*x - 4 , 'a')
n = K.degree()
B = [1, a, a^2, a^3, 1/2*a^4 - 1/2*a^2]  # Basis for the ring of integers

def elems_to_matrix_power (l):
    M = [x.list() + [0] * (n - len(x.list())) for x in l]
    return Matrix(M).transpose()

def elems_to_basis (x):
    return (elems_to_matrix_power(B)).inverse() * (elems_to_matrix_power (x))

# From a m x n - matrix of lists (of length r), constructs a r x (mn) matrix that has those lists as columns: 
# The entry (i,j) corresponds to the ni + j column 
def matrix_of_matrix_lists (M, m, n):
    L = []
    for i in range(m):
        for j in range(n):
            L = L + [M[i][j]]
    M = Matrix(L).transpose()
    return(M)
            
    
O = K.ring_of_integers()

ideal_gens = [11, a^2 + 3*a + 3]  

I = O.ideal(ideal_gens)

# Compute a Z-basis for the ideal I
Z_basis = I.basis()

M = Matrix(K, len(ideal_gens), n)
MB = [[[] for j in range(n)] for i in range(len(ideal_gens))]

for i in range(len(ideal_gens)):
    for j in range(n):
        M[i, j] = (ideal_gens[i]) * B[j]
        MB[i][j] = (elems_to_basis ([M[i,j]])).list()

Mland = Matrix(ZZ, matrix_of_matrix_lists(MB,len(ideal_gens), n))
MAux, Echel = Mland.transpose().echelon_form(transformation=True)
faux = [() for k in range(n)]
for k in range(n):
    faux[k] = Echel.transpose() * (MAux.transpose()).solve_right(vector(elems_to_basis([Z_basis[k]])))
    
f = [[[[] for j in range(n)] for i in range(len(ideal_gens))] for k in range(n)]
for k in range(n):
    for i in range(len(ideal_gens)):
        for j in range(n):
            f[k][i][j] = faux[k][n * i + j]
           

g = [[[[] for k in range(n)] for j in range(n)] for i in range(len(ideal_gens))]
for i in range(len(ideal_gens)):
    for j in range(n) :
        g[i][j] = (elems_to_basis(Z_basis)).solve_right(vector(MB[i][j]))

w = [[] for i in range(n)]  
v = [[] for i in range(len(ideal_gens))]

for i in range(len(ideal_gens)):
    v[i] = (elems_to_basis ([ideal_gens[i]])).list()
for j in range(n):
    w[j] = (elems_to_basis ([Z_basis[j]])).list()
    
    
print(v)
print(w)
print(Mland)

print(f)
```
